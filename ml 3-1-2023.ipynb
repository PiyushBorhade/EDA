# SAMPLING TECHNIQUE


## 1: Oversampler
## 2: Undersampler
## 3: SMOTE

X should be in 2D format

y should be in 1D format

```python
# 1: over sampler
# 2: under sampler
# 3: smote (will create new records by itself)
```

```python
X = df.iloc[:,:30]                       # first method

#X = df.drop('Class',axis=1)               # second method 

y = df['Class']
```

```python
X
```

```python
y.value_counts()
```

```python
!pip install imblearn
```

# Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler()
```

```python
X_ros,y_ros = ros.fit_resample(X,y)
```

```python
X_ros.shape
```

```python
y_ros.shape
```

```python
y_ros.value_counts()                     # it converts sample into 50%-50%
```

# Under Sampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler()
```

```python
X_rus,y_rus = rus.fit_resample(X,y)
```

```python
X_rus.shape
```

```python
y_rus.shape
```

```python
y_rus.value_counts()
```

# SMOTE (Synthetic minority oversapling technique)

```python
from imblearn.over_sampling import SMOTE
```

```python
smt = SMOTE()
```

```python
X_smt,y_smt = smt.fit_resample(X,y)
```

```python
X_smt.shape
```

```python
y_smt.shape
```

```python
y_smt.value_counts()
```
